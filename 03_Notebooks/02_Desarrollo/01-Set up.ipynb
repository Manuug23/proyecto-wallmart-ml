### Se unen los dataframes features y stores

```python
df = features.merge(right = stores, how = 'left', on = 'Store').reset_index(drop = True).set_index('Date')
df.info()
```

### Se unen los dataframes df con train (para obtener la var Dept y la target)

```python
import pandas as pd

# Asegurarnos de que la columna 'Date' en el DataFrame 'train' esté en formato de fecha
train['Date'] = pd.to_datetime(train['Date'])

# Convertir el índice de 'df' a una columna para facilitar el merge
df.reset_index(inplace=True)

# Realizar la combinación de los DataFrames utilizando 'merge'
df_combined = pd.merge(df, train[['Store', 'Dept', 'Date', 'Weekly_Sales']], 
                       on=['Store', 'Date'], how='left')

```

```python
df = df_combined.copy()
```

```python
df.set_index('Date', inplace = True)
```

### Extraer y reservar el dataset de validación

```python
val = df.sort_index().loc['2012-07-26':'2013-07-26']
val
```

```python
nombre_fichero_validacion = 'validacion.csv'
ruta_completa = path + '/02_Datos/02_Validacion/' + nombre_fichero_validacion

val.to_csv(ruta_completa)
```

### Extraer y  guardarel dataset de trabajo

```python
trabajo = df.drop(val.index)
trabajo
```

```python
nombre_fichero_trabajo = 'trabajo.csv'

ruta_completa = path + '/02_Datos/03_Trabajo/' + nombre_fichero_trabajo

trabajo.to_csv(ruta_completa)
```
